```python

import pandas as pd
import tabula
import os

#diretório que está o PDF
diretorio_arquivo = r'C:\Users\U003659\Desktop'

#Percorre todos os arquivos na pasta especificada
for name_file in os.listdir(diretorio_arquivo):
    # Se o nome do arquivo for pdf e 'Citi 02' está no nome do arquivo, fará a leitura do arquivo especificado
    if name_file.endswith('.pdf') and 'Citi 02' in name_file:
        arquivo = tabula.read_pdf(os.path.join(diretorio_arquivo, name_file), pages='all')

# Verifica se a variável "arquivo" contém elementos
if arquivo:

    try: 

        # Define o primeiro bloco do PDF
        tabela1 = arquivo[0]

        # Define a linha 6 do bloco do PDF como cabeçalho
        tabela1.columns = tabela1.iloc[6]

        # Define a linha 8 do bloco do PDF como dados das colunas
        tabela1 = tabela1[8:]

        #Modificação da coluna "período de data de data" para menter apenas a primeira data
        tabela1['Período de Data de Data'] = tabela1['Período de Data de Data'].apply(lambda x: str(x).split()[0])

        #Identifica valores com traço na coluna "valor da transação" e cria uma nova coluna "débito"
        tabela1['Débito'] = tabela1['Valor da Transação'].apply(lambda x: x.replace('-', '') if '-' in str(x) else 0)

        #Converte valores em "valor da transação" para débito (caso tenha valores negativos)
        tabela1['Valor da Transação'] = tabela1['Valor da Transação'].apply(lambda x: 0 if '-' in str(x) else x)

        # Exibir o arquivo após as alterações
        tabela1.to_excel(r'C:\Users\U003659\Desktop\Extrato_citi_02.xlsx')
    except:

        # Define o primeiro bloco do PDF
        tabela1 = arquivo[1]
        
        # Define a linha 6 do bloco do PDF como cabeçalho
        tabela1.columns = tabela1.iloc[6]
    
        # Define a linha 8 do bloco do PDF como dados das colunas
        tabela1 = tabela1[8:]
    
        #Modificação da coluna "período de data de data" para menter apenas a primeira data
        tabela1['Período de Data de Data'] = tabela1['Período de Data de Data'].apply(lambda x: str(x).split()[0])
    
        #Identifica valores com traço na coluna "valor da transação" e cria uma nova coluna "débito"
        tabela1['Débito'] = tabela1['Valor da Transação'].apply(lambda x: x.replace('-', '') if '-' in str(x) else 0)
    
        #Converte valores em "valor da transação" para débito (caso tenha valores negativos)
        tabela1['Valor da Transação'] = tabela1['Valor da Transação'].apply(lambda x: 0 if '-' in str(x) else x)
    
        # Exibir o arquivo após as alterações
        tabela1.to_excel(r'C:\Users\U003659\Desktop\Extrato_citi_02.xlsx')
else:
    print("PDF vazio")
```

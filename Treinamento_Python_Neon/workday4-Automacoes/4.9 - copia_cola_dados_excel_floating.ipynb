```python
import openpyxl

# Faz a leitura da planilha base
planilha_base = openpyxl.load_workbook(r'C:\Users\U003659\Desktop\dados_extraidos.xlsx')
aba_base = planilha_base.active

# Faz a leitura da planilha destino
planilha_destino = openpyxl.load_workbook(r'C:\Users\U003659\Desktop\Controle Resgates Citi LFT_Julho 2024.xlsx')
aba_destino = planilha_destino.active


''' Parte do código que identifica a primeira linha vazia da planilha destino '''

# Encontrar a primeira linha vazia na planilha destino da coluna 1
linha_destino = 1
while aba_destino.cell(row=linha_destino, column=1).value is not None:
    linha_destino += 1

# Encontrar a primeira linha vazia na planilha destino da coluna 4
linha_destino_2 = 1
while aba_destino.cell(row=linha_destino_2, column=4).value is not None:
    linha_destino_2 += 1

# Encontrar a primeira linha vazia na planilha destino da coluna 5
linha_destino_3 = 1
while aba_destino.cell(row=linha_destino_3, column=5).value is not None:
    linha_destino_3 += 1

# Encontrar a primeira linha vazia na planilha destino da coluna 6
linha_destino_4 = 1
while aba_destino.cell(row=linha_destino_4, column=6).value is not None:
    linha_destino_4 += 1

# Encontrar a primeira linha vazia na planilha destino da coluna 7
linha_destino_5 = 1
while aba_destino.cell(row=linha_destino_5, column=7).value is not None:
    linha_destino_5 += 1

# Encontrar a primeira linha vazia na planilha destino da coluna 8
linha_destino_6 = 1
while aba_destino.cell(row=linha_destino_6, column=8).value is not None:
    linha_destino_6 += 1

# Encontrar a primeira linha vazia na planilha destino da coluna 9
linha_destino_7 = 1
while aba_destino.cell(row=linha_destino_7, column=9).value is not None:
    linha_destino_7 += 1

# Encontrar a primeira linha vazia na planilha destino da coluna 2
linha_destino_8 = 1
while aba_destino.cell(row=linha_destino_8, column=2).value is not None:
    linha_destino_8 += 1

# Encontrar a primeira linha vazia na planilha destino da coluna 3
linha_destino_9 = 1
while aba_destino.cell(row=linha_destino_9, column=3).value is not None:
    linha_destino_9 += 1


''' Parte do código que encontra a linha do cabeçalho da planilha base '''

# Encontrar a linha onde está o cabeçalho "Data liquidação"
linha_cabecalho = 1
while aba_base.cell(row=linha_cabecalho, column=6).value != "Data Liquidação":
    linha_cabecalho += 1

# Encontrar a linha onde está o cabeçalho "Valor Bruto (BRL)"
linha_cabecalho_2 = 1
while aba_base.cell(row=linha_cabecalho_2, column=11).value != "Valor Bruto (BRL)":
    linha_cabecalho_2 += 1

# Encontrar a linha onde está o cabeçalho "Valor Líquido (BRL)"
linha_cabecalho_3 = 1
while aba_base.cell(row=linha_cabecalho_3, column=14).value != "Valor Líquido (BRL)":
    linha_cabecalho_3 += 1

# Encontrar a linha onde está o cabeçalho "Imposto de Renda Retido na Fonte (BRL)"
linha_cabecalho_4 = 1
while aba_base.cell(row=linha_cabecalho_4, column=13).value != "Imposto de Renda Retido na Fonte (BRL)":
    linha_cabecalho_4 += 1

# Encontrar a linha onde está o cabeçalho "Número da Operação"
linha_cabecalho_5 = 1
while aba_base.cell(row=linha_cabecalho_5, column=1).value != "Número da Operação":
    linha_cabecalho_5 += 1

# Encontrar a linha onde está o cabeçalho "Quantidade de Títulos Liquidados"
linha_cabecalho_6 = 1
while aba_base.cell(row=linha_cabecalho_6, column=10).value != "Quantidade de Títulos Liquidados":
    linha_cabecalho_6 += 1

# Encontrar a linha onde está o cabeçalho "Preço Unitário Corrigido (BRL)"
linha_cabecalho_7 = 1
while aba_base.cell(row=linha_cabecalho_7, column=9).value != "Preço Unitário Corrigido (BRL)":
    linha_cabecalho_7 += 1

# Encontrar a linha onde está o cabeçalho "Conta Citi"
linha_cabecalho_8 = 1
while aba_base.cell(row=linha_cabecalho_7, column=15).value != "Conta Citi":
    linha_cabecalho_8 += 1



''' Parte do código que pega as informações da planilha base e copia para a planilha destino '''

# Iterar sobre as linhas abaixo do cabeçalho da sexta coluna da planilha base e copiar para a primeira coluna da planilha destino
for linha, valor_celula in enumerate(aba_base.iter_rows(min_row=linha_cabecalho + 1, min_col=6, max_col=6, values_only=True), start=1):
    for valor in valor_celula:
        aba_destino.cell(row=linha_destino, column=1, value=valor)
        linha_destino += 1

# Iterar sobre as linhas abaixo do cabeçalho da décima primeira coluna da planilha base e copiar para a quarta coluna da planilha destino
for linha_2, valor_celula_2 in enumerate(aba_base.iter_rows(min_row=linha_cabecalho_2 + 1, min_col=11, max_col=11, values_only=True), start=1):
    for valor in valor_celula_2:
        aba_destino.cell(row=linha_destino_2, column=4, value=valor)
        linha_destino_2 += 1

# Iterar sobre as linhas abaixo do cabeçalho da décima quarta coluna da planilha base e copiar para a quinta coluna da planilha destino
for linha_3, valor_celula_3 in enumerate(aba_base.iter_rows(min_row=linha_cabecalho_3 + 1, min_col=14, max_col=14, values_only=True), start=1):
    for valor in valor_celula_3:
        aba_destino.cell(row=linha_destino_3, column=5, value=valor)
        linha_destino_3 += 1

# Iterar sobre as linhas abaixo do cabeçalho da décima terceira coluna da planilha base e copiar para a sexta coluna da planilha destino
for linha_4, valor_celula_4 in enumerate(aba_base.iter_rows(min_row=linha_cabecalho_4 + 1, min_col=13, max_col=13, values_only=True), start=1):
    for valor in valor_celula_4:
        aba_destino.cell(row=linha_destino_4, column=6, value=valor)
        linha_destino_4 += 1

# Iterar sobre as linhas abaixo do cabeçalho da primeira coluna da planilha base e copiar para a sétima coluna da planilha destino
for linha_5, valor_celula_5 in enumerate(aba_base.iter_rows(min_row=linha_cabecalho_5 + 1, min_col=1, max_col=1, values_only=True), start=1):
    for valor in valor_celula_5:
        aba_destino.cell(row=linha_destino_5, column=7, value=valor)
        linha_destino_5 += 1      

# Iterar sobre as linhas abaixo do cabeçalho da décima coluna da planilha base e copiar para a oitava coluna da planilha destino
for linha_6, valor_celula_6 in enumerate(aba_base.iter_rows(min_row=linha_cabecalho_6 + 1, min_col=10, max_col=10, values_only=True), start=1):
    for valor in valor_celula_6:
        aba_destino.cell(row=linha_destino_6, column=8, value=valor)
        linha_destino_6 += 1  

# Iterar sobre as linhas abaixo do cabeçalho da nona coluna da planilha base e copiar para a oitava nona da planilha destino
for linha_7, valor_celula_7 in enumerate(aba_base.iter_rows(min_row=linha_cabecalho_7 + 1, min_col=9, max_col=9, values_only=True), start=1):
    for valor in valor_celula_7:
        aba_destino.cell(row=linha_destino_7, column=9, value=valor)
        linha_destino_7 += 1  

# Iterar sobre as linhas abaixo do cabeçalho da décima quinta coluna da planilha base e copiar para a terceira da planilha destino
for linha_8, valor_celula_8 in enumerate(aba_base.iter_rows(min_row=linha_cabecalho_8 + 1, min_col=15, max_col=15, values_only=True), start=1):
    for valor in valor_celula_8:
        aba_destino.cell(row=linha_destino_8, column=3, value=valor)
        linha_destino_8 += 1

# Iterar sobre as linhas abaixo do cabeçalho da décima quinta coluna da planilha base e copiar para a segunda da planilha destino
for linha_9, valor_celula_9 in enumerate(aba_base.iter_rows(min_row=linha_cabecalho_8 + 1, min_col=15, max_col=15, values_only=True), start=1):
    for valor in valor_celula_8:
        aba_destino.cell(row=linha_destino_9, column=2, value=valor)
        linha_destino_9 += 1  

# Salvar as alterações na planilha destino
planilha_destino.save('Controle Resgates Citi LFT_Julho 2024.xlsx')
```

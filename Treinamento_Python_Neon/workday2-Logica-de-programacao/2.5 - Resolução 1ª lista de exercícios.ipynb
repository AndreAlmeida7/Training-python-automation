```python
# Exercício 1

# Calculadora em Python
# Desenvolva uma calculadora em Python com tudo que você aprendeu nos capítulos até aqui no curso. 
# A solução será apresentada no próximo capítulo!

print("******************* Calculadora em Python *******************")



while True: 
    
    numero_1 = input('Digite um número: ')
    numero_2 = input('Digite um número: ')
    operador = input('Digite um operador matemático (+-/*): ')


    numeros_validos = None

    try:
        primeiro_algarismo = float(numero_1)
        segundo_algarismo = float(numero_2)
        numeros_validos = True
    except:
        numeros_validos = None

    if numeros_validos is None:
        print('Um dos número são inválidos')
        continue

    operadores_validos = '+-/*'

    if operador not in operadores_validos:
        print('O operador matemático digitado, não é válido')
        continue


    if operador == '*':
        resultado = primeiro_algarismo * segundo_algarismo
        print(resultado)
    elif operador == '+':
        resultado = primeiro_algarismo + segundo_algarismo
        print(resultado)
    elif operador == '-':
        resultado = primeiro_algarismo - segundo_algarismo
        print(resultado)
    elif operador == '/':
        resultado = primeiro_algarismo / segundo_algarismo
        print(resultado) 

    sair = input('Quer sair? Aperte a sigla s: ')    

    if sair == 's': 
        break 
```

```python
#Exercício 2

'''
Escreva um programa que solicite ao usuário um número inteiro positivo e 
faça uma contagem regressiva desse número até zero, exibindo cada número na tela.
'''

# Solicita ao usuário um número inteiro positivo
numero = int(input("Digite um número inteiro positivo: "))

# Verifica se o número é positivo
if numero > 0:
    # Realiza a contagem regressiva
    for i in range(numero, -1, -1):
        print(i)
else:
    print("Por favor, digite um número inteiro positivo.")

```

```python
#Exercício 3

'''Escreva um programa que solicite ao usuário um número inteiro e exiba a tabuada desse número de 1 a 10.'''

# Solicita ao usuário um número inteiro
numero = int(input("Digite um número inteiro: "))

# Exibe a tabuada de 1 a 10
print(f"Tabuada do {numero}:")
for i in range(1, 11):
    print(f"{numero} x {i} = {numero * i}")
```

```python
#Exercício 4

'''
Escreva um programa que solicite ao usuário uma quantidade de números a serem inseridos.
Leia esses números e calcule a média deles.
'''

# Solicita ao usuário a quantidade de números a serem inseridos
quantidade = int(input("Digite a quantidade de números que deseja inserir: "))

# Inicializa a variável que armazenará a soma dos números
soma = 0

# Lê os números e calcula a soma
for i in range(quantidade):
    numero = float(input(f"Digite o número {i + 1}: "))
    soma += numero

# Calcula a média
if quantidade > 0:
    media = soma / quantidade
    print(f"A média dos números inseridos é: {media:.2f}")
else:
    print("A quantidade de números deve ser maior que zero.")
```

```python
#Exercício 5

'''
Faça um jogo para o usuário adivinhar qual a palavra secreta.
Você vai propor uma palavra secreta qualquer e irá dar a possibilidade para o usuário digitar apenas uma letra.
Quanto o usuário digitar uma letra, você vai conferir se a letra digitada está na palavra secreta.

Se a letra digitada estiver na palavra secreta; exiba a letra;
Se a letra digitada não estiver na palavra secreta; exiba *.

Faça a contagem de tentativas do seu usuário.
Se o usuário digitar várias letras peça a ele digitar apenas uma letra.

'''

palavra_secreta = 'elefante'
numero_tentativas = 0
letra_acertada = '' #Variável de letras que estão corretas na palavra secreta


while True:
    #Pede ao jogador a digitar uma letra
    letra_digitada = input('Digite uma letra para começar a jogar: ')
    numero_tentativas = numero_tentativas + 1

    if len(letra_digitada) > 1:
        print('Digite apenas uma letra!')
        continue

    if letra_digitada in palavra_secreta:
        letra_acertada = letra_acertada + letra_digitada

    palavra_formatada = '' #Variável para formatar a palavra formada no 'for'.
    for letra_secreta in palavra_secreta: #Para cada letra secreta que estiver dentro da palavra secreta fará:
        if letra_secreta in letra_acertada: #Lógica para saber se a letra secreta está nas letras acertadas
            palavra_formatada += letra_secreta #se estiver, acrescenta a letra secreta na variável da palavra formatada
        else:
            palavra_formatada += '*' #se não estiver, a a palavra formatada exibirá asteriscos
    print(palavra_formatada) #mostra a palavra formatada com asteriscos e, se houver, uma letra correta.

    if palavra_formatada == palavra_secreta:
        print(f'Você ganhou o jogo parabéns! A palavra secreta é {palavra_secreta}! ') #Mostrará que ganhou e a palavra secreta
        print(f'Você acertou a palavra em {numero_tentativas} tentativas!') #Mostrará o número de tentativas
        sair = input('Deseja sair do jogo? Aperta a tecla "s": ')
        if sair == 's':
            break
```

```python
#Exercício 6 

'''
Faça um programa que peça ao usuário para digitar um número inteiro,
informe se este número é par ou ímpar. Caso o usuário não digite um número
inteiro, informe que não é um número inteiro.

par -> é divisível por 2 -> resto da divisão por 2 é zero. 

'''

numero = input('Digite um número: ')


lista_pares = [0,2,4,6,8]


if numero.isdigit():
    if int(numero[-1]) in lista_pares:
        print(f'O número {numero} é par')
    else:
        print(f'O número {numero} é ímpar')
else:
    print(f'O número {numero} não é inteiro')
```

```python
#Exercício 7

'''
Exercício
peça ao usuário para digitar seu nome
Peça ao usuário digitar sua idade
Se nome e idade forem digitados:
    Exiba:
        Seu nome é {nome}
        Seu nome contém (ou não) espaços
        Seu nome tem {n} letras
        A primeira letra do seu nome é {letra}
        A última letra do seu nome é {letra}
Se nada for digitado em nome ou idade:

exiba " Desculpe, você deixou campos vazios."

'''


nome = input('digite seu nome: ')
idade = input('digite sua idade: ')


if not nome or not idade:
    print("Desculpe, você deixou os campos vazios.")
else:
    print(f'Seu nome é {nome}')
    if " " in nome:
        print('Seu nome contém espaços')
    else:
        print('Seu nome não tem espaço')
    print('Seu nome tem ' f'{len(nome)} ' 'letras')
    print('A segunda letra do seu nome é ' f'{nome[1]}')
    print(f'A penultima letra do seu nome é {nome[-2]}')
```

```python
#Exercício 8

"""
Faça um programa que pergunte a hora ao usuário e, baseando-se no horário 
descrito, exiba a saudação apropriada. Ex. 
Bom dia 0-11, Boa tarde 12-17 e Boa noite 18-23.

"""

entrada = input("Escreva a hora desejada: ")

if entrada.isdigit():
    hora = int(entrada)
    if hora >= 0 and hora <= 11:
        print(f'Bom dia! Agora são {hora} horas da manhã')
    elif hora >= 12 and hora <= 17:
        print(f'Boa tarde! Agora são {hora} horas da tarde')
    elif hora >= 18 and hora <= 23:
        print(f'Boa noite! Agora são {hora} horas da noite')

else:
    print('VOcê não digitou a hora corretamente, tente novamente')
```
